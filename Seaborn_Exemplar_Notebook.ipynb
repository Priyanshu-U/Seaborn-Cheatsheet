Seaborn Data Visualisation

```python
#Importing Libraries 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Using the inbuilt dataset

```python
#total bill vs top dataset
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
 #setting a theme for the plot
 sns.set_theme()
```

Relational Plotting

```python
#relational plotting using seasborn 
sns.relplot(data=tips,x='total_bill',y='tip',col='sex',style='smoker',hue='smoker',size='size')
```

```python
#house price dataset
from sklearn.datasets import load_boston
house_boston=load_boston()

house=pd.DataFrame(house_boston.data, columns=house_boston.feature_names)
house['PRICE']=house_boston.target

house.head()
```

Distribution Plot

```python
sns.distplot(house['PRICE'])
```

Heat Map

```python
#construction of a heat map
correlation=house.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True, fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```
